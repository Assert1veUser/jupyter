```python
def randPeriod(X):
    result = []
    n = len(X)
    for i in range(0, n):
        ele = X[i]
        for j in range(i, n):
            if ((ele == X[j]) & (i != j)):
                result.append(j - 1)
                result.append(i)
                result.append(j)
                return result
    result.append(-1)
    result.append(-1)
    result.append(-1)
    return result
```

```python
test_1 = randPeriod(rParamsArr_e1)
test_2 = randPeriod(rParamsArr_e2)
test_3 = randPeriod(rParamsArr_e3)
test_4 = randPeriod(rParamsArr_e4)

print("TEST_1 = ", test_1)
print()
print("TEST_2 = ", test_2)
print()
print("TEST_3 = ", test_3)
print()
print("TEST_4 = ", test_4)
```

```python
import math
def getFreqDistr(rParamsArr, A, B, IntervalsCount):
    Freq = []
    dY = (B - A)/IntervalsCount
    for i in range(0, IntervalsCount):
        Freq.append(0)
    for j in range(0, len(rParamsArr)):
        Yc = rParamsArr[j]
        fN = int(math.floor(Yc/dY))
        Freq[fN] +=1
    for i in range(0, IntervalsCount):
        Freq[i] = Freq[i]/(len(rParamsArr) * dY)
    return Freq
```

```python
import matplotlib.pyplot as plt
K = 10
resX = []
for k in range(0, K):
    resX.append(((B - A)/K) * (0.5 + k))
resY = getFreqDistr(rParamsArr_e1, A, B, K)
plt.bar(resX, resY)

X_Pirsona = 0
for i in range(0, K):
    X_Pirsona = X_Pirsona + ((1/K - resY[i]) ** 2)/resY[i]
print("    N=10^2 = ", X_Pirsona)
```
