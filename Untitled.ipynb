```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import preprocessing
from scipy.cluster.hierarchy import dendrogram
from scipy.cluster.hierarchy import linkage

count = {'count_room': [2, 1, 1, 1, 1, 1, 1, 1, 1],
'count_rack': [39, 21, 21, 20, 18, 15, 15, 13, 12]}
df = pd.DataFrame(data=count,
index=['Client 8 (0)', 'Client 7 (1)','Client 3 (2)','Client 1 (3)',
'Client 2 (4)','Client 9 (5)','Client 6 (6)', 'Client 4 (7)', 'Client 5 (8)'])

scaler = preprocessing.MinMaxScaler().fit(df.to_numpy())
scaled_data = scaler.transform(df.to_numpy())

distance_matrix_ward = linkage(scaled_data, method='ward',
metric='euclidean')
fig = plt.figure(figsize=(18,9))
fig.patch.set_facecolor('white')
R = dendrogram(distance_matrix_ward,
leaf_font_size=12)
plt.show()
```
